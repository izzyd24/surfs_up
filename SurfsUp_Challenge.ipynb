```python
# Dependencies
import numpy as np

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import pandas as pd
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# I am combining steps 2 and 3 as one
# 2/3 Write a query that filters the Measurement table to retrieve the temperatures for the month of June + as a list
surfing_june = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==6).all()
print(surfing_june)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
surf_june_df = pd.DataFrame(surfing_june, columns=['date', 'June Temperatures'])
surf_june_df.sample()
```

```python
# show without index to make less confusing
surf_june_df = surf_june_df.sort_index()
print(surf_june_df.to_string(index=False))
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
surf_june_df.describe()
```

```python
# plot for better understanding
surf_june_df.plot()
```

## D2: Determine the Summary Statistics for December

```python
# combining steps 6 and 7 as one
# 6/7. Write a query that filters the Measurement table to retrieve the temperatures for the month of December + to list
surfing_dec = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==12).all()
print(surfing_dec)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
surf_dec_df = pd.DataFrame(surfing_dec, columns=['date', 'Dec. Temperatures'])
surf_dec_df.sample()
```

```python
# show without index to make less confusing
surf_dec_df = surf_dec_df.sort_index()
print(surf_dec_df.to_string(index=False))
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
surf_dec_df.describe()
```
